```python
id_like_cols = [c for c in df.columns if any(k in c.lower() for k in ['id','name','email','phone'])]
id_like_cols
```

```python
df_masked = df.copy()
if 'name' in df_masked.columns:
    df_masked['name_masked'] = df_masked['name'].str[0] + '***'
df_masked.head()
```

```python
df_model = df.drop(columns=id_like_cols, errors='ignore').copy()
df_model.head()
```

```python
# example: enforce gpa range 0-4, age >= 0
df_model.loc[(df_model['gpa'] < 0) | (df_model['gpa'] > 4), 'gpa'] = np.nan
df_model.loc[df_model['age'] < 0, 'age'] = np.nan
df_model.isnull().sum()
```

## Qn / Ans (Tutor Answers)

**Qn 1:** Which column is clearly identifier-like?  
**Ans:** `name` is identifier-like and should not be used as predictor.

**Qn 2:** Why mask data even in class practice?  
**Ans:** To build habit of protecting personally identifying information.

**Qn 3:** Is every existing column suitable for modelling?  
**Ans:** No. Remove identifiers, leaked outcomes, or irrelevant fields first.
