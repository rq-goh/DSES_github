```python
df = pd.read_csv('hiring_data.csv')
df.head()
```

```python
num_cols = ['age','gpa','years_of_exp']
df[num_cols].describe()
```

```python
df[num_cols].skew()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))
sns.histplot(df['years_of_exp'], kde=True, ax=ax[0])
ax[0].set_title('years_of_exp (raw)')
df['years_of_exp_log'] = np.log1p(df['years_of_exp'])
sns.histplot(df['years_of_exp_log'], kde=True, ax=ax[1])
ax[1].set_title('years_of_exp_log')
plt.tight_layout()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='gpa', y='years_of_exp', hue='hired')
plt.title('GPA vs Experience by hired')
plt.show()
```

```python
df['gpa_x_exp'] = df['gpa'] * df['years_of_exp']
df[['gpa','years_of_exp','gpa_x_exp']].head()
```

## Qn / Ans (Tutor Answers)

**Qn 1:** Is `years_of_exp` skewed?  
**Ans:** Yes, it is positively skewed in the raw view; log transform makes it more balanced.

**Qn 2:** Why compare before/after transform?  
**Ans:** To check if transformation improves shape and interpretability.

**Qn 3:** Give one valid engineered feature.  
**Ans:** `gpa_x_exp = gpa * years_of_exp` as a simple interaction-style feature.
