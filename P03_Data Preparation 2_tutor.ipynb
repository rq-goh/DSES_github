## Lets Load

```python
import pandas as pd
import numpy as np
import os
from google.colab import drive
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer, KNNImputer
```

```python
drive.mount('/content/drive')
os.chdir('/content/drive/MyDrive/DSES')
os.getcwd()
```

```python
df = pd.read_csv('loan_data.csv')
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()
num_cols, cat_cols
```

```python
df_mean = df.copy()
df_median = df.copy()
df_mode = df.copy()

mean_imp = SimpleImputer(strategy='mean')
median_imp = SimpleImputer(strategy='median')
mode_imp = SimpleImputer(strategy='most_frequent')

df_mean[num_cols] = mean_imp.fit_transform(df_mean[num_cols])
df_median[num_cols] = median_imp.fit_transform(df_median[num_cols])
df_mode[cat_cols] = mode_imp.fit_transform(df_mode[cat_cols])

df_mean.isnull().sum(), df_mode.isnull().sum()
```

```python
df_knn = df.copy()
knn_imp = KNNImputer(n_neighbors=3)
df_knn[num_cols] = knn_imp.fit_transform(df_knn[num_cols])
df_knn.head()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))
sns.histplot(df['loan_amt'], kde=True, ax=ax[0])
ax[0].set_title('Original loan_amt')
sns.histplot(df_mean['loan_amt'], kde=True, ax=ax[1])
ax[1].set_title('After mean imputation')
plt.tight_layout()
```

## Qn / Ans (Tutor Answers)

**Qn 1:** Which column has missing values?  
**Ans:** `sector` has about **91** missing values in this dataset.

**Qn 2:** Should mean/median be used on categorical columns?  
**Ans:** No. Use mode/most_frequent for categorical columns.

**Qn 3:** Why compare methods instead of choosing one directly?  
**Ans:** Different methods change distributions differently and can affect modelling later.
