### Talking point

In this dataset, `sector` contains missing values (about 91). Numeric columns are complete.

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()
num_cols, cat_cols
```

## Step 2 - Build clean copies for comparison

```python
df_mean = df.copy()
df_median = df.copy()
df_mode = df.copy()
df_knn = df.copy()
```

## Step 3 - Imputation (SimpleImputer)

```python
mean_imp = SimpleImputer(strategy='mean')
median_imp = SimpleImputer(strategy='median')
mode_imp = SimpleImputer(strategy='most_frequent')

df_mean[num_cols] = mean_imp.fit_transform(df_mean[num_cols])
df_median[num_cols] = median_imp.fit_transform(df_median[num_cols])
df_mode[cat_cols] = mode_imp.fit_transform(df_mode[cat_cols])

df_mean.isnull().sum(), df_mode.isnull().sum()
```

## Step 4 - KNN imputation for numeric columns

```python
knn_imp = KNNImputer(n_neighbors=3)
df_knn[num_cols] = knn_imp.fit_transform(df_knn[num_cols])
df_knn[num_cols].head()
```

## Step 5 - Compare effect on distribution

```python
fig, ax = plt.subplots(1,3, figsize=(15,4))
sns.histplot(df['loan_amt'], kde=True, ax=ax[0])
ax[0].set_title('Original loan_amt')
sns.histplot(df_mean['loan_amt'], kde=True, ax=ax[1])
ax[1].set_title('Mean-imputed loan_amt')
sns.histplot(df_median['loan_amt'], kde=True, ax=ax[2])
ax[2].set_title('Median-imputed loan_amt')
plt.tight_layout()
```

## Guided activity: method justification (10-15 mins)

Ask students to justify:
1. Why `most_frequent` is reasonable for `sector`
2. Why mean vs median can behave differently when distribution is skewed
3. Why KNN is not automatically better

## Qn / Ans (Tutor answers with dataset-grounded points)

**Qn 1:** Which column has missing values in this dataset?  
**Ans:** `sector` has missing values (about **91** rows).

**Qn 2:** Should we use `mean` on `sector`?  
**Ans:** No, `sector` is categorical; use `most_frequent` or other category-aware method.

**Qn 3:** Why compare mean and median even if missing count is low in numeric columns?  
**Ans:** Because distribution sensitivity differs; median is often more robust to outliers/skew.

**Qn 4:** What is the key exam habit from this practical?  
**Ans:** State a reason for your imputation choice, not only code.

## Common mistakes to highlight

- Filling every column with one method
- Forgetting to separate numeric and categorical columns
- Fitting imputer on wrong subset
- No before/after inspection
