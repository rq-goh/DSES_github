Organized Structure: <br>

# 1.0 Pandas DF: Using pd to read csv to create DF

```python
import pandas as pd
from io import StringIO

# Create a CSV string with the specific values on separate rows.
csv_data = """A
1,0
2,NA
3,null
4,None
5,
6,nil"""

# Read the CSV data.
df = pd.read_csv(StringIO(csv_data))

# Display the DataFrame.
print(df)

# Check the type of each cell in column A.
for index, value in df['A'].items():
    print(f"Row {index}: Value = {value}, Type = {type(value)}")
```

```python
import pandas as pd
import numpy as np

list1 = [0, 'NA', 'null', None,np.nan, 'nil']
list1_df = pd.DataFrame(list1, columns=['A'])
print(list1_df)

for index, value in list1_df['A'].items():
    print(f"Row {index}: Value = {value}, Type = {type(value)}")
```

```python
#lets try another dataset
import pandas as pd

last_names = ['Connor', 'Connor', 'Reese', 'Doe', 'Doe','Doe','Doey']
first_names = ['Sarah', 'John', 'Kyle', 'John','John','Jack', 'Jill']
age = [23,25,21,20,19,20,20]
df = pd.DataFrame({
  'first_name': first_names,
  'last_name': last_names,
  'age': age
})
df
```

```python
#print unique values
print(df['first_name'].unique())
print(df['last_name'].unique())
```

```python
#print count per unique value
print(df['first_name'].value_counts())
print(df['last_name'].value_counts())
```

```python
#lets point to rows with John
(df['first_name']=='John')
```

```python
#average age and count by first_name
df.groupby('first_name')['age'].agg(['mean','count'])
```

```python
#lets point to rows with John
df[df['first_name']=='John']
```

```python
#lets point to rows with Doe
df[df['last_name']=='Doe']
```

```python
df[df['age']>=20]
```

```python
df[df['age']!=20]
```
