```python
df = pd.read_csv('hiring_data.csv')
df.head()
```

```python
df.columns
```

```python
target = 'hired'
id_like_cols = [c for c in df.columns if any(k in c.lower() for k in ['id','name','email','phone'])]
possible_leakage = []
remove_cols = list(set(id_like_cols + possible_leakage))
remove_cols
```

```python
X = df.drop(columns=remove_cols + [target], errors='ignore')
y = df[target]
X.head(), y.head()
```

```python
X_num = X.select_dtypes(include=[np.number]).columns.tolist()
X_cat = X.select_dtypes(exclude=[np.number]).columns.tolist()
X_num, X_cat
```
