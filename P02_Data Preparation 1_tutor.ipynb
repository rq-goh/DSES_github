## Step 4 - Label/condition-based selection with loc

```python
df.loc[df['beer_servings'] > 200, ['country','beer_servings','continent']].head(15)
```

```python
df.loc[(df['wine_servings'] > 100) & (df['continent'] == 'EU'), ['country','wine_servings','continent']].head(15)
```

```python
df.loc[df['total_litres_of_pure_alcohol'] >= 10, ['country','total_litres_of_pure_alcohol']].head(15)
```

## Timed mini-task (15-20 mins)

Ask students to complete these in order:
1. Show last 8 rows
2. Show rows 20-35, columns `country` to `wine_servings`
3. Countries in Asia (`AS`) with `spirit_servings > 100`
4. Top 5 countries by total litres of pure alcohol
5. Count how many countries have `beer_servings == 0`

```python
# Tutor answers
ans1 = df.tail(8)
ans2 = df.iloc[20:36, 0:4]
ans3 = df.loc[(df['continent']=='AS') & (df['spirit_servings']>100), ['country','spirit_servings','continent']]
ans4 = df.sort_values('total_litres_of_pure_alcohol', ascending=False).head(5)[['country','total_litres_of_pure_alcohol']]
ans5 = (df['beer_servings'] == 0).sum()

ans1.head(), ans2.head(), ans3.head(), ans4, ans5
```

## Common mistakes to highlight

- Confusing `iloc` (position) vs `loc` (labels/conditions)
- Using single brackets when multiple columns are needed
- Writing long code before checking `columns` and `dtypes`
- Not verifying shape after filtering
