## Indexing and slicing

```python
df_simple.iloc[0]
```

```python
df_simple.iloc[1:4, 0:3]
```

```python
df_simple.loc[df_simple['score'] >= 80]
```

```python
df_simple[['name','score']]
```

## Qn / Ans (Tutor Answers)

**Qn 1:** Why not inspect full dataframe every time?  
**Ans:** Large data wastes time. Use targeted slices to inspect exactly what you need.

**Qn 2:** What is the difference between `.iloc` and `.loc`?  
**Ans:** `.iloc` uses integer positions, `.loc` uses labels/conditions.

**Qn 3:** Give one ICA benefit from this practical.  
**Ans:** Faster row/column selection means quicker debugging and fewer setup mistakes in timed tasks.
