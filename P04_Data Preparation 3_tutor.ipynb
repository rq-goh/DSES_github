### Dataset context to explain

- Balanced origin split close to USA vs non-USA

## Step 2 - Univariate EDA

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Price'], kde=True)
plt.title('Distribution of Price')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Horsepower'])
plt.title('Horsepower boxplot (outlier check)')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='Type', order=df['Type'].value_counts().index)
plt.xticks(rotation=40)
plt.title('Count of car Type')
plt.show()
```

## Step 3 - Bivariate / multivariate EDA

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Horsepower', y='Price', hue='Origin')
plt.title('Horsepower vs Price by Origin')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Weight', y='MPG.city', hue='Type')
plt.title('Weight vs MPG.city by Type')
plt.show()
```

```python
pd.crosstab(df['Type'], df['Origin'])
```

```python
corr = df[num_cols].corr()
plt.figure(figsize=(11,8))
sns.heatmap(corr, cmap='coolwarm')
plt.title('Numeric correlation heatmap')
plt.show()
```

## Step 4 - Feature suggestion from EDA

```python
df['hp_per_mpg_city'] = df['Horsepower'] / df['MPG.city']
df[['Horsepower','MPG.city','hp_per_mpg_city']].head()
```

```python
df[['Price','Horsepower','Weight','MPG.city','hp_per_mpg_city']].corr()
```

## Qn / Ans (Tutor answers with dataset-grounded observations)

**Qn 1:** Any major missing values issue here?  
**Ans:** In this version of `Cars93.csv`, missing counts are generally 0 across columns.

**Qn 2:** Most common `Type` categories?  
**Ans:** `Midsize` (~22) and `Small` (~21) are the most frequent.

**Qn 3:** What does `Horsepower vs Price` suggest?  
**Ans:** Clear positive tendency: higher horsepower often links to higher price.

**Qn 4:** Why create `hp_per_mpg_city`?  
**Ans:** It combines performance and fuel-efficiency perspective, potentially helpful for later modelling.
