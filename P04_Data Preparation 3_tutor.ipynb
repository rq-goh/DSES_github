```python
df = pd.read_csv('Cars93.csv')
df.head()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns
cat_cols = df.select_dtypes(exclude=[np.number]).columns
len(num_cols), len(cat_cols)
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Price'], kde=True)
plt.title('Price distribution')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Horsepower'])
plt.title('Horsepower boxplot')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(data=df, x='Type', order=df['Type'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Car type counts')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Horsepower', y='Price', hue='Origin')
plt.title('Horsepower vs Price')
plt.show()
```

```python
pd.crosstab(df['Type'], df['Origin'])
```

```python
corr = df.select_dtypes(include=[np.number]).corr()
plt.figure(figsize=(10,7))
sns.heatmap(corr, cmap='coolwarm')
plt.title('Correlation heatmap')
plt.show()
```

```python
df['hp_per_mpg_city'] = df['Horsepower'] / df['MPG.city']
df[['Horsepower','MPG.city','hp_per_mpg_city']].head()
```
